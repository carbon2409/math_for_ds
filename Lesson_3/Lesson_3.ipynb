```python
#Задание 3 Окружность
import matplotlib.pyplot as plt
import math as m
import numpy as np
%matplotlib inline
r = 2
x = np.linspace(-r, r, 1000)
y = []
y_ = []
for i in x:
    #Находим значения у по формуле: y**2 = r**2 - x**2
    y.append(m.sqrt(r**2 - i**2))
    #Находим отрицательные y
    y_.append(-m.sqrt(r**2 - i**2))
    
    
    


plt.plot(x, y)
plt.plot(x, y_)
plt.show()
#График сжатый, наверное потому, что фигура прямоугольная. (Подзабыл, как настраивать axes)
```

```python
#Эллипс
a = 4
b = 3
x = np.linspace(-a, a, 1000)
y = []
y_ = []
for i in x:
    y.append(m.sqrt(1 - i**2/a**2)*b)
    y_.append(-m.sqrt(1 - i**2/a**2)*b)

plt.plot(x, y)
plt.plot(x, y_)
plt.grid()
plt.show()
#Cнова проблемки с фигурой :(
```

```python
#Гипербола
x = []
y = []
a = 2
b = 5
for i in range(1000*2*a):
    x_ = a + i/1000
    x.append(x_)
    y.append(m.sqrt(x_**2/a**2 - 1) * b)

plt.plot(x, y)

plt.grid()
plt.show()
```

```python
#Задание 5 Две параллельные плоскости
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
axes = Axes3D(fig)
x = np.arange(-10, 10, 1)
y = np.arange(-10, 10, 1)
x, y = np.meshgrid(x, y)
z = 2*x + 3*y
z2 = 2*x + 3*y + 100
axes.plot_surface(x, y, z)
axes.plot_surface(x, y, z2)
show()
```

```python
#Две параллельные фигуры 2-го порядка
fig = figure()
axes = Axes3D(fig)
x = np.arange(-10, 10, 1)
y = np.arange(-10, 10, 1)
x, y = np.meshgrid(x, y)
z = (x**4 + y**4)
z2 = - (x**4 + y**4)
axes.plot_surface(x, y, z)
axes.plot_surface(x, y, z2)
show()
```
